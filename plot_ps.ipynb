```python
import numpy as np
import matplotlib.pyplot as plt

archivo = "Summary_Evolution_Cosmic_Perturbations_N_7_EO78H_TEST_3.dat"

datos = np.loadtxt(archivo)


N = datos[:, 3]
L = datos[:, 4]
kvar = datos[:, 10]/(np.exp(datos[:,3])*datos[:,2])
kcom = datos[:, 10]
z = datos[:, 12]
zeta = kcom**3 * L**2 / (2 * np.pi**2 * z**2)




plt.figure(figsize =(8,5))


plt.yscale('log')

plt.plot(np.log(kvar), np.abs(zeta), marker = ".", color = 'r')
plt.grid()


plt.show()
```
